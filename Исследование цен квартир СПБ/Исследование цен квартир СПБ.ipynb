### Шаг 2. Предобработка данных

```python
data.duplicated().sum()
```

#### Медианное значение высоты потолков - 2.65м, что является реалистичной величиной, произведем замену пропущеных значений.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data['ceiling_height'].isnull().sum()
```

#### Медианное значение количества этажей - 9, что соотетствует средней этажности построек в городах РФ.

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'].isnull().sum()
```

#### Медианное значение жилой площади 30кв.м, что примерно соответствует однокомнатной квартире

```python
data['living_area'] = data['living_area'].fillna(data['living_area'].median())
data['living_area'].isnull().sum()
```

#### Пропущеные значения апартаментов примем за False

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['floors_total'].isnull().sum()
```

#### Медианное значение площади кухни -9 кв.м., что так же выглядит вполне реалистично

```python
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())
data['kitchen_area'].isnull().sum()
```

####  Пропущенные значения в графе 'Балконы' примем за их отсутствие

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isnull().sum()
```

#### Пропущенные значения ближайших аэропортов заменим нулями, т.к. в основном пропуски характерны для деревень, поселков и т.д. где отсутствие аэропортов гарантировано

```python
data['airports_nearest'] = data['airports_nearest'].fillna(0)
data['airports_nearest'].isnull().sum()
```

#### Медианное значение близости к центру - 13 км, что вполне укладывается в реалии. Заменим пропущеные значения медианой

```python
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(data['cityCenters_nearest'].median())
data['cityCenters_nearest'].isnull().sum()
```

#### Пропущенные значения парков поблизости примем за 0.

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['parks_around3000'].isnull().sum()
```

#### Так как парков в радиусе 3км может не быть, значит и расстояние до ближайшего парка пропущено. Заменим пропуски 0.

```python
data['parks_nearest'] = data['parks_nearest'].fillna(0)
data['parks_nearest'].isnull().sum()
```

#### Отсутствие прудов в радиусе 3км - заменим на 0.

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data['ponds_around3000'].isnull().sum()
```

#### Пропущенные значения расстояния до ближайшего пруда так же заменим 0

```python
data['ponds_nearest'] = data['ponds_nearest'].fillna(0)
data['ponds_nearest'].isnull().sum()
```

#### Пропущенные значения говорят нам, что объявление было удалено в день публикации, поэтоум заменим кол-во дней на 0

```python
data['days_exposition'] = data['days_exposition'].fillna(0)
data['days_exposition'].isnull().sum()
```

#### Приведем значения следующих столбцов к целочисленному виду для удобства работы, т.к. точность до десятых долей в данных столбцах не играет никакой роли, в отличае от площадей кухни, квартиры и тд.

```python
data['last_price'] = data['last_price'].astype('int64')
data['floors_total'] = data['floors_total'].astype('int64')
data['balcony'] = data['balcony'].astype('int64')
data['airports_nearest'] = data['airports_nearest'].astype('int64')
data['cityCenters_nearest'] = data['cityCenters_nearest'].astype('int64')
data['parks_around3000'] = data['parks_around3000'].astype('int64')
data['parks_nearest'] = data['parks_nearest'].astype('int64')
data['ponds_around3000'] = data['ponds_around3000'].astype('int64')
data['ponds_nearest'] = data['ponds_nearest'].astype('int64')
data['days_exposition'] = data['days_exposition'].astype('int64')
```

#### Пропущенных значений в столбце населенного пункта - 49, что составляет 0.21% от общего числа объявлений. Данные потери считаю допустимыми и предлагаю удалить.

### Шаг 3. Посчитаем и дабавим в таблицу данные

#### Посчитаем цену квадратного метра

```python
data['price_metr'] = data['last_price'] / data['total_area']
data['price_metr'] = data['price_metr'].astype('int64')
```

#### Переведем колонки first_day_exposition к формату дат при помощи to_datetime.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

#### Категоризация квартир по 3 группам этажей (первый, последний, другой)

```python
def floor_category (row):
    total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return ('первый этаж')
    if floor != 1 and floor != total:
        return ('другой этаж')
    if floor == total:
        return ('последний этаж')
    
row_values = ['floors_total',  'floors']
row_columns = ['floors_total',  'floors']
row = pd.Series(data=row_values, index=row_columns)
data['floor_category'] = data.apply(floor_category, axis=1)
```

#### Соотношение жилой площади к общей

```python
data['living_to_all'] = data['living_area'] / data['total_area']
```

#### Соотношение кухни к общей площади

```python
data['kitchen_to_all'] = data['kitchen_area'] / data['total_area']
```
